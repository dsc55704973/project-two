```python
import requests
import pandas as pd
from config import api_key
import json
```

### Pull Brewery Data from Yelp

```python


url = 'https://api.yelp.com/v3/businesses/search'
headers = {
        'Authorization': 'Bearer {}'.format(api_key),
    }

url_params = {
                "categories": 'breweries, All',
                "location":"Denver",
                "state": "CO",
                "limit": 50
            }
response = requests.get(url, headers=headers, params=url_params).json()
# print(response)

response
```

#### Brewery Data into Useable Dataframe

```python
df_brew = pd.DataFrame.from_dict(response['businesses'])
df_brew
df_brew_2=pd.concat([df_brew.drop('coordinates', axis=1), pd.DataFrame(df_brew['coordinates'].tolist())], axis=1)
df_brew_3=pd.concat([df_brew_2.drop('location', axis=1), pd.DataFrame(df_brew_2['location'].tolist())], axis=1)
df_brew_3.head()
```

### Pull Distillery Data from Yelp

```python
url = 'https://api.yelp.com/v3/businesses/search'
headers = {
        'Authorization': 'Bearer {}'.format(api_key),
    }

url_params_3 = {
                "categories": 'distilleries, All',
                "location":"Denver",
                "state": "CO",
                "limit": 50
            }
response_3 = requests.get(url, headers=headers, params=url_params_3).json()
response_3
```

```python
df_denver=pd.concat([df_brew_3, df_wine_3, df_whiskey_3])
df_denver.head()
```
